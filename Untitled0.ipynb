```python
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import LabelEncoder

df = pd.read_csv("/content/claims_clean_format_with_anomalies.csv")

# Convert date
df["date"] = pd.to_datetime(df["date"])
df["hour"] = df["date"].dt.hour
df["day_of_week"] = df["date"].dt.dayofweek
df["is_weekend"] = df["day_of_week"].isin([5, 6]).astype(int)

# Aggregated behavioral features
df["claims_by_patient"] = df.groupby("patientName")["claimId"].transform("count")
df["claims_by_provider"] = df.groupby("provider")["claimId"].transform("count")
df["avg_amount_by_patient"] = df.groupby("patientName")["amount"].transform("mean")
df["avg_amount_by_provider"] = df.groupby("provider")["amount"].transform("mean")
df["avg_amount_by_diagnosis"] = df.groupby("diagnosisCode")["amount"].transform("mean")

first_dates = df.groupby("patientName")["date"].transform("min")
df["days_since_first_claim"] = (df["date"] - first_dates).dt.days

# A very strong anomaly signal: deviation from patient's mean
df["amount_diff_from_patient_mean"] = df["amount"] - df["avg_amount_by_patient"]

# Encode categoricals
le_patient = LabelEncoder()
le_provider = LabelEncoder()
le_diag = LabelEncoder()

df["patient_encoded"] = le_patient.fit_transform(df["patientName"])
df["provider_encoded"] = le_provider.fit_transform(df["provider"])
df["diagnosis_encoded"] = le_diag.fit_transform(df["diagnosisCode"])

# Features for model
features = [
    "amount",
    "claims_by_patient",
    "claims_by_provider",
    "avg_amount_by_patient",
    "avg_amount_by_provider",
    "avg_amount_by_diagnosis",
    "days_since_first_claim",
    "amount_diff_from_patient_mean",
    "day_of_week",
    "is_weekend",
    "patient_encoded",
    "provider_encoded",
    "diagnosis_encoded"
]

X = df[features]

# Train Isolation Forest
iso = IsolationForest(
    n_estimators=200,
    contamination='auto',   # assume 5% anomalies in data
    random_state=42
)
iso.fit(X)

# Predict
df["anomaly_pred"] = iso.predict(X)   # 1 = normal, -1 = anomaly
df["anomaly_pred"] = df["anomaly_pred"].map({1: 0, -1: 1})  # make 1 = anomaly

print("Predicted anomalies:", df["anomaly_pred"].sum())
```
